```python
# Replace inf values with NaN
df.replace([np.inf, -np.inf], np.nan, inplace=True)

# Drop NaN values to avoid warnings
df.dropna(inplace=True)

# Convert 'Class' to string (ensures proper color mapping)
df['Class'] = df['Class'].astype(str)

# Sample a subset of the data for better performance
df_sample = df.sample(n=1000, random_state=42)  # Adjust sample size as needed
```

```python
# Generate the pairplot

sns.pairplot(df,hue='Class',palette='Set1')
```

## Train Test and Split

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('Class',axis=1)
y = df['Class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

## Decision Tress

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier(criterion='entropy', random_state=0)
```

```python
dtree.fit(X_train,y_train)
```

## Prediction and Evaluation

```python
predictions = dtree.predict(X_test)
```

```python
predictions
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
from sklearn import tree
plt.figure(figsize=(20,25))
tree.plot_tree(dtree,feature_names=X.columns,class_names=['Class-1', 'Class-0'],rounded=True, # Rounded node edges
          filled=True, # Adds color according to class
          proportion=True
        )
plt.show()
```
